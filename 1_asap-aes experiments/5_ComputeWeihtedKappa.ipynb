# Dataset1_ASAP_AES

```python
import pandas as pd
from scipy import stats
import os
# Folder path containing the CSV files
folder_path = r'E:\Research work\assessment checker research\asap-aes\results_gemini'
```

## Gemini

```python
from sklearn.metrics import cohen_kappa_score
# Loop over each file from EssaySet1 to EssaySet8
for i in range(1, 7):
    file_name = f'EssaySet{i}_assessment_results_by_rubric.csv'
    file_path = os.path.join(folder_path, file_name)
    
    # Read the CSV file
    if os.path.exists(file_path):
        data = pd.read_csv(file_path)
        if i==1:
            data['domain1_score']= data['domain1_score']//2
        if 'domain1_Point_gemini' in data.columns:            
            data['domain1_Score_gemini'] = data['domain1_Score_gemini'].fillna(data['domain1_Point_gemini'])
            
        # print(f"Checking missing values in EssaySet{i}...")
        # print(f"EssaySet{i}.. Empty records", data['domain1_Score_gemini'].isnull().sum())
        df_cleaned = data[['domain1_score', 'domain1_Score_gemini']].dropna()


        # Calculate the weighted kappa score
        kappa_score = cohen_kappa_score(df_cleaned['domain1_score'], df_cleaned['domain1_Score_gemini'], weights='quadratic')
        print(f"EssaySet{i}  Domain1; Weighted Kappa Score: {round(kappa_score,2)}")
        print()
        if i==2:
            data = pd.read_csv(file_path)
            # print(data.columns)
            if 'domain2_Point_gemini' in data.columns:            
                data['domain2_Score_gemini'] = data['domain2_Score_gemini'].fillna(data['domain2_Point_gemini'])       
            df_cleaned = data[['domain2_score', 'domain2_Score_gemini']].dropna()
            kappa_score = cohen_kappa_score(df_cleaned['domain2_score'], df_cleaned['domain2_Score_gemini'], weights='quadratic')
            print(f"EssaySet{i}  Domain2; Weighted Kappa Score: {round(kappa_score,2)}")
            print()
```

```python
# Dataset2_Learning Agency
import pandas as pd

# Load the data from the provided Excel file
file_path = 'learning-agency-lab-automated-essay-scoring-2/results_gpt/Holistic_assessment_results_by_rubric.csv'
data = pd.read_csv(file_path)

# Display the first few rows of the dataframe
data.head()
data['Holistic Score_GPT'] = data['Holistic Score_GPT'].fillna(data['Holistic Rating_GPT'])
    
print(data[['score', 'Holistic Score_GPT']].isnull().sum())
data = data[['score', 'Holistic Score_GPT']].dropna()
data['Holistic Score_GPT'] = data['Holistic Score_GPT'].round()


# Calculate the weighted kappa score
kappa_score = cohen_kappa_score(data['score'], data['Holistic Score_GPT'], weights='quadratic')

print(f"Weighted Kappa Score: {round(kappa_score,2)}")

```
